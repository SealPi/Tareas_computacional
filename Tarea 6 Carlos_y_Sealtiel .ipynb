hint: Aplica iterativamente el algoritmo reduciendo el grado de la raiz, hasta obtener una raiz cuadrada que puedas obtener con el algoritmo babilónico!

````julia
function Raiz_babilónica(x, a = x, n=1, ϵ = 0.001)
    contador = 0
    while abs(a^n - x) > ϵ && contador <1000
        contador += 1
        a = (n-1)/n * a + x/(n*a^(n-1))
        @show a
    end
    return a
end 
````

**2**. Fácil: En el video dimos argumentos heurísticos de por qué los puntos fijos atractivos deben tener su derivada en el intervalo $(-1, 1)$. Formaliza estos argumentos para demostrar que es así. 

Concideremos un mapeo lineal $x_{n+1}=\alpha x_n$, el mapeo define una sucesión
geométrica. Entonces, el $n$-ésimo iterado será:

\begin{equation}
x_n = x_0 \alpha^n
\end{equation}

Es fácil ver que , si $0\ge\alpha<1$, el límite cuando $n\to\infty$ es 0, el punto
fijo. De igual manera, si $-1<\alpha<0$ el límite será también el punto fijo. Y el punto fijo es **linealmente estable**.

Por otra parte, si $|\alpha|>1$, el límite cuando $n\to\infty$ es $\infty$, es decir,
los iterados se alejan del punto fijo. En este caso el punto
es **linealmente inestable**.

--------------------------------------------------------------------------------------------------------------------------------

El análisis que hemos desarrollado aquí lo hicimos para mapeos lineales de la forma
$x_{n+1} = F_\alpha(x_n)=\alpha x_n$, pero es útil más allá de los mapeos lineales.

Ahora cualquier mapeo $F(x)$, al rededor de un punto fijo lo podemos expresar como $F(x_* + \delta)$
que al expandirlo en **Taylot** hasta segundo orden, tenemos:

\begin{equation}
F(x_* + \delta) = F(x_*) + \delta F'(x_*) + \cal{O}(\delta^2),
\end{equation}

Si $x_*$ es punto fijo entonces se cumple $f(x_*)=0$ y como la perturvación es pequeña, entonces podemos despreciar el término cuadrático, por lo que:

$$F(x)=F(x_* + \delta) \approx \delta F'(x_*)$$

Y es un mapeo lineal en $\delta$!!, donde el equivalente de la
pendiente $\alpha$ utilizada arriba, es $F'(x_*)$. Por lo que la estabilidad esta dada por la derivada en el punto fijo y además cumple las desigualdades

$$-1<F'(x_*)<0$$
$$0\ge F'(x_*)<1$$

6. Fácil: Un modelo de población razonable que incluye una taza de natalidad (- mortalidad) y una taza de migración fuera una comunidad es: 
$$ N(t) = N_o e^{\lambda t} + \frac{v}{\lambda} (e^{\lambda t} -1)$$
    Donde $v$ es una constante de migración.

    De una polación de aves se sabe que había $N_0 = 10^6$ individuos inicialmente, que $v = 435,000$ migraron el primer año y $1,564,000$ se registraron al final del primer año. ¿Cuál es la taza de natalidad $\lambda$ de esta población? Utiliza el métod de la Bisección o de Newton para resolver la ecuación. 

````julia
##### Función de la que se desea encontrar la raíz (lambda)
function f_pob(x,N₀,N₁,v)
        return exp(x)*(N₀ + v/x) - v/x - N₁
end
##### Derivada de la función 
function df_pob(x,N₀,N₁,v)
        return exp(x)*(N₀ + v/x - v/x^2) + v/x^2
end
##### tolerancia de error, diferencia entre "aproximaciones sucesivas"
epsilon = 1e-15
##### Aproximación inicial de la raíz
x_i = 0.5
##### Error inicial, no importa su valor numérico, pero tiene que ser mayor que la tolerancia (epsilon)
error = 0.5
##### Contador
i=0

##### Valor de los parámetros
N₀ = 1e+6
N₁ = 1.564e+6
v = 435e+3

##### Condición sobre la diferencia entre aproximaciones sucesivas
while error>epsilon
    x_n = x_i - f_pob(x_i,N₀,N₁,v)/df_pob(x_i,N₀,N₁,v)
    error = abs(x_n - x_i)
    x_i = x_n
    i = i+1
    ###### límite de pasos para salir del bucle aunque no se haya encontrado una buena aproximación a la raíz
    if i==1e+6
        println("No se ha llegado a la precisión deseada")
        break
    end
end
println("Aproximación de la raíz:")
println(x_i)
println("Error aproximado:")
println(error)
println("Número de pasos:")
println(i)
````

Finalmente, en la bici puede mantener una velocidad de 30km/h. Calcula cuál es la ruta que debe seguir el deportista para llegar en el menor tiempo posible a la meta y grafica dicha trayectoria. 



````julia
function α(x)
    return x/(14*sqrt(x^2 +16))
end
function γ(x)
    return 4*30*x/sqrt(14^2*(x^2 +16)-30^2*x^2)
end
function β(x)
    return 10-γ(x)-x
end
function f_tri(x)
    return 2/21+α(x)-(25/21)*β(x)/sqrt(25*(β(x))^2+0.25*21)
end
````

````julia
function bisec(f,a,b)
    if f(a)*f(b)>=0
        println("Intervalo inválido")
        return 0
    end
    tol = 1e-10
    epsilon = 1e-9
    error = 1
    i=0
    c = (a+b)/2
    while error>epsilon
        if abs(f(c))<tol
            return c
        else
            if f(a)*f(c)<0
                b=c
            else 
                if f(c)*f(b)<0
                    a=c
                else 
                    return 0
                end
            end
        end
        c = (a+b)/2
        i = i+1
        error = (b-a)/2
        if i==1e+9
            println("Pasos límite alcanzados")
            return c
        end
    end
    return c
end                        
````

````julia
x1 = bisec(f_tri,1,2)
x2 = β(x1)
x3 = γ(x1)
println("Tierra: ",x1,"[km]","\n","Agua: ", x2,"[km]","\n","Bici: ", x3,"[km]")
````

Si integramos esta densidad entre $a$ y $b$ obtenemos la probabilidad de encontrar la partícula en el intervalo $[a,b]$. Buscaremos en el rango de valores $[\mu -x, \mu +x]$. ¿Cuanto vale $x$ para que la probabilidad de encontrar la partícula sea $0.5$?

    Nota: puedes resolver esto por cualquier método, Punto fijo, Bisección, Newton o intervalos.

````julia
function p_normal(y)
    return exp(-(y-5)^2/(2*1.2^2))/(1.2*sqrt(2*pi))
end
function integral_simpson(f, a,b;n = 100)
    Δx = (b-a)/n
    sum([(f(x)+f(x+Δx)+4*f(x+Δx/2)) for x ∈ a:Δx:(b-Δx)].*Δx/6)
end
function f_8(z)
    return (integral_simpson(p_normal,5,5+z) -0.25)
end
function df_8(z)
    return (p_normal(5+z))
end
````

````julia
##### tolerancia de error, diferencia entre "aproximaciones sucesivas"
epsilon = 1e-15
##### Aproximación inicial de la raíz
x_i = 2
##### Error inicial, no importa su valor numérico, pero tiene que ser mayor que la tolerancia (epsilon)
error = 0.5
##### Contador
i=0
tol = 1e-10

##### Condición sobre la diferencia entre aproximaciones sucesivas
while error>epsilon
    if abs(f_8(x_i))<tol
        break
    end
    if abs(df_8(x_i))<tol
        break
    end
    x_n = x_i - f_8(x_i)/df_8(x_i)
    error = abs(x_n - x_i)
    x_i = x_n
    i = i+1
    #println(x_i)
    ###### límite de pasos para salir del bucle aunque no se haya encontrado una buena aproximación a la raíz
    if i==1e+6
        println("No se ha llegado a la precisión deseada")
        break
    end
end
println("Aproximación de la raíz:")
println(x_i)
println("Error aproximado:")
println(error)
println("Número de pasos:")
println(i)
````

9. Haz una función que obtenga la intersección entre un plano inclinado (dado un punto por el que pasa el plano y el vector normal al plano) y una superficie de revolución al rededor del eje $z$. Para esto primero haz la función que dado un punto por el que pasa el plano, el vector normal y las coordenadas $x,y$, obtenga la coordenada $z$ del plano y la función que dada una función $z(x)$ y las coordenadas $x,y$ arroje la coordenada $z$ de la superficie de revolución de $z(x)$ al rededor del eje $z$. 

    Prueba tu función con la recta $z(x) = x$ y diferentes planos inclinados (deberías obtener las secciones cónicas). 

Hint: Para obtener la intersección, debes resolver un sistema de 3 ecuaciones no lineales, una de las ecuaciones es la superficie de revolución, otra es el plano inclinado y la última es el plano $z = const$, para diferentes constantes. 


````julia
function z(x)
    return 2*x
end
function Z(x,y)
    return z(sqrt(x^2 + y^2))
end
function z_p(n,P,x,y)
    if n[3]==0
        println("Plano horizontal")
    end
    return (P[3]/n[3] -(x-P[1])*P[1]/n[3] - (y-P[2])*P[2]/n[3])
end
````

````julia
a = -4
b = 4
c = -4
d = 4

X = a:.01:b
Y = c:.01:d

n = [0,3,1]
P = [0,1,2]
epsilon = 1e-2
XX=[]
YY=[]
ZZ=[]
for i in 1:length(X)
    for j in 1:length(Y)
        if abs(Z(X[i],Y[j]) - z_p(n,P,X[i],Y[j]))<epsilon
            push!(XX,X[i])
            push!(YY,Y[j])
            push!(ZZ,Z(X[i],Y[j]))
        end
    end
end
````
