**1**. Fácil: Variante del algoritmo para raices de orden n: Para calcular la raiz cuadrada cambiamos un cuadrado de área $x$ por un rectángulo. Generaliza esta idea para hacer un algoritmo que calcule la raiz $n$-esima de un número cambiando de un hipercubo a un hiperprisma. La función debe tomar como argumentos el número $x$ al que se quiere sacar la raiz, una adivinanza inicial $a_0$ y el grado de la raiz $n$. 

hint: Aplica iterativamente el algoritmo reduciendo el grado de la raiz, hasta obtener una raiz cuadrada que puedas obtener con el algoritmo babilónico!

```julia
function Raiz_babilónica(x, a = x, n=1, ϵ = 0.001)
    contador = 0
    while abs(a^n - x) > ϵ && contador <1000
        contador += 1
        a = (n-1)/n * a + x/(n*a^(n-1))
        @show a
    end
    return a
end 
```

```julia
Raiz_babilónica(8,1,17)
```

**2**. Fácil: En el video dimos argumentos heurísticos de por qué los puntos fijos atractivos deben tener su derivada en el intervalo $(-1, 1)$. Formaliza estos argumentos para demostrar que es así. 

Concideremos un mapeo lineal $x_{n+1}=\alpha x_n$, el mapeo define una sucesión
geométrica. Entonces, el $n$-ésimo iterado será:

\begin{equation}
x_n = x_0 \alpha^n
\end{equation}

Es fácil ver que , si $0\ge\alpha<1$, el límite cuando $n\to\infty$ es 0, el punto
fijo. De igual manera, si $-1<\alpha<0$ el límite será también el punto fijo. Y el punto fijo es **linealmente estable**.

Por otra parte, si $|\alpha|>1$, el límite cuando $n\to\infty$ es $\infty$, es decir,
los iterados se alejan del punto fijo. En este caso el punto
es **linealmente inestable**.

--------------------------------------------------------------------------------------------------------------------------------

El análisis que hemos desarrollado aquí lo hicimos para mapeos lineales de la forma
$x_{n+1} = F_\alpha(x_n)=\alpha x_n$, pero es útil más allá de los mapeos lineales.

Ahora cualquier mapeo $F(x)$, al rededor de un punto fijo lo podemos expresar como $F(x_* + \delta)$
que al expandirlo en **Taylot** hasta segundo orden, tenemos:

\begin{equation}
F(x_* + \delta) = F(x_*) + \delta F'(x_*) + \cal{O}(\delta^2),
\end{equation}

Si $x_*$ es punto fijo entonces se cumple $f(x_*)=0$ y como la perturvación es pequeña, entonces podemos despreciar el término cuadrático, por lo que:

$$F(x)=F(x_* + \delta) \approx \delta F'(x_*)$$

Y es un mapeo lineal en $\delta$!!, donde el equivalente de la
pendiente $\alpha$ utilizada arriba, es $F'(x_*)$. Por lo que la estabilidad esta dada por la derivada en el punto fijo y además cumple las desigualdades

$$-1<F'(x_*)<0$$
$$0\ge F'(x_*)<1$$
